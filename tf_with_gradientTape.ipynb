```python
import tensorflow as tf

#Load and preprocess the MNIST dataset
(x_train, y_train), (x_test,y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test= x_train /255.0, x_test/255.0 # Normalize to [0,1]
x_train 
```

```python
x_train = x_train.reshape(-1, 28* 28)
x_test = x_test.reshape(-1, 28 * 28)

# Convert labels to one-hot encoding
y_train_onehot = tf.keras.utils.to_categorical(y_train, num_classes = 10)
y_test_onehot = tf.keras.utils.to_categorical(y_test, num_classes=10)
```

```python
class SimpleNN(tf.keras.Model):#·Äí·ÄÆ·Äú·Ä≠·ÄØ·Äõ·Ä±·Ä∏·Äê·Ä¨·ÄÄüëâ TensorFlow ·Äõ·Ä≤·Ä∑ Model class ·ÄÄ·Ä≠·ÄØ ·Ä°·Äô·ÄΩ·Ä±·ÄÜ·ÄÄ·Ä∫·ÄÅ·Ä∂·Äï·Äº·ÄÆ·Ä∏üëâ ·ÄÄ·Ä≠·ÄØ·Äö·Ä∫·Äï·Ä≠·ÄØ·ÄÑ·Ä∫ Neural Network ·Äê·ÄÖ·Ä∫·ÄÅ·ÄØ ·Äê·Ää·Ä∫·ÄÜ·Ä±·Ä¨·ÄÄ·Ä∫·Äê·Ä¨·Äï·Ä´·Åã
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')
        self.dense2 = tf.keras.layers.Dense(10, activation='softmax')

    def call(self,inputs):
        x = self.dense1(inputs)
        return self.dense2(x)
```

```python

model = SimpleNN()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
loss_fn = tf.keras.losses.CategoricalCrossentropy()
```

```python
epochs = 3
batch_size = 32
num_batches = len(x_train) // batch_size

print("num_batches : ",num_batches)
```

```python
1000//32
```

```python
model.trainable_variables
```

```python
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")

    # new models creations
    # set weights from best model

    for i in range(num_batches):
        # Get a batch of data
        start = i * batch_size

        end = start + batch_size
        x_batch = x_train[start:end]
        y_batch = y_train_onehot[start: end]

        with tf.GradientTape() as tape:#üëâ Model ·Äõ·Ä≤·Ä∑ computation (·Äê·ÄΩ·ÄÄ·Ä∫·ÄÅ·Äª·ÄÄ·Ä∫·Äô·Äæ·ÄØ·Äê·ÄΩ·Ä±) ·ÄÄ·Ä≠·ÄØ ·Äô·Äæ·Äê·Ä∫·Äê·Äô·Ä∫·Ä∏·Äê·ÄÑ·Ä∫·Äë·Ä¨·Ä∏·Äê·Ä≤·Ä∑ recorder ·Äê·ÄÖ·Ä∫·ÄÅ·ÄØ·Äú·Ä≠·ÄØ·Äï·Ä´·Åã
            predictions = model(x_batch,training=True) #Forward pass
            loss = loss_fn(y_batch,predictions) # Compute loss

        gradients =tape.gradient(loss, model.trainable_variables)#üëâ Loss ·ÄÄ·Ä≠·ÄØ ·Äú·Äª·Ä±·Ä¨·Ä∑·ÄÖ·Ä±·Äñ·Ä≠·ÄØ·Ä∑üëâ Weight ·Äê·ÄÖ·Ä∫·ÄÅ·ÄØ·ÄÅ·Äª·ÄÑ·Ä∫·Ä∏ ·Äò·Äö·Ä∫·Äú·Ä±·Ä¨·ÄÄ·Ä∫ ·Äï·Äº·Ä±·Ä¨·ÄÑ·Ä∫·Ä∏·Äõ·Äô·Äú·Ä≤ ·Äê·ÄΩ·ÄÄ·Ä∫·Äê·Ä¨·Åã·Äí·Ä´·ÄÄ Backpropagation ·Äï·Ä´·Åã

        optimizer.apply_gradients(zip(gradients,model.trainable_variables)) #Update weights

        if i % 200 == 0: #Batch 200 ·Äê·ÄÖ·Ä∫·ÄÅ·Ä´·Äê·Ä≠·ÄØ·ÄÑ·Ä∫·Ä∏ Loss ·Äï·Äº·Äê·Äö·Ä∫·Åã Training ·Äô·Äæ·Ä¨ loss ·Äú·Äª·Ä±·Ä¨·Ä∑·Äû·ÄΩ·Ä¨·Ä∏·Äõ·ÄÑ·Ä∫ model ·ÄÄ learn ·Äú·ÄØ·Äï·Ä∫·Äî·Ä±·Äê·Ä¨·Äï·Ä´·Åã
            print(f"Batch {i}/{num_batches}, Loss: {loss.numpy():.4f}")
```

```python
#Evaluate the model
test_loss = loss_fn(y_test_onehot,model(x_test))
test_accuracy = tf.keras.metrics.categorical_accuracy(
    y_test_onehot,model(x_test))#Prediction ·Äë·Ä≤·ÄÄ probability ·Ä°·Äô·Äª·Ä¨·Ä∏·ÄÜ·ÄØ·Ä∂·Ä∏ class ·ÄÄ·Ä≠·ÄØ ·Äö·Ä∞·Äô·Äö·Ä∫True label ·Äî·Ä≤·Ä∑ ·Äê·Ä∞·Äô·Äê·Ä∞ ·ÄÖ·ÄÖ·Ä∫·Äô·Äö·Ä∫·Äê·Ä∞·Äõ·ÄÑ·Ä∫ 1, ·Äô·Äê·Ä∞·Äõ·ÄÑ·Ä∫ 0
print(f"Test Loss: {test_loss.numpy():.4f}, Test Accuracy: {tf.reduce_mean(test_accuracy).numpy():.4f}")
```
